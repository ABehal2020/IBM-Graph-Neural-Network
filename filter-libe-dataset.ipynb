```python
import pandas as pd
import json

df = pd.read_json("libe.json")
```

```python
df["Keep"] = True
df["Number of non hydrogen atoms"] = None
```

```python
# loop over all rows in data frame specifically looking at the composition column
# if any element is heavier than Calcium, throw this row out
# also keep a counter for the number of atoms in the molecule corresponding to the row you are looking at
# however, do not increment the counter if the value corresponds to the number of hydrogen atoms
# throw out the row if the counter is greater than 10

print(len(df))
```

```python
# first 20 elements of the periodic table
validElements = ["H", "He", "Li", "Be", "B", "C", "N", "O", "F", "Ne", "Na", "Mg", "Al", "Si", "P", "S", "Cl", "Ar", "K", "Ca"]

print(len(validElements))
```

```python
for i in range(len(df)):
    composition = dict(df.loc[i]["composition"])
    # number of atoms that are not hydrogen
    numAtoms = 0
    for key in composition:
        if key not in validElements:
            df.at[i, "Keep"] = False
        if key != "H":
            numAtoms += (int)(composition[key])
    df.at[i, "Number of non hydrogen atoms"] = numAtoms
    if numAtoms > 10:
        df.at[i, "Keep"] = False
```
